## Plotting Energy Consumption Over Several Months

```python
# Sample data for energy consumption over a 6 months (in MWh)
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun']
energy_consumption = [1200, 1300, 1100, 1500, 1400, 1600]

# Create a line plot
plt.plot(months, energy_consumption, marker='o', color='b', linestyle='--')

# Add labels and title
plt.title('Energy consumption over 6 months')
plt.xlabel('Months')
plt.ylabel('Energy consumption (MWh)')
plt.show()
```

## Bar Chart on Comparing Energy Consumption by Different Renewable Sources

```python
# Sample data for energy consumption by different renewable sources
energy_sources = ['Solar', 'Wind', 'Hydropower', 'Biomass']
energy_values = [1200, 3400, 2900, 2500]

# Create a bar chart
plt.bar(energy_sources, energy_values, color='blue')

# Add labels and title
plt.title('Energy consumption by renewable sources')
plt.xlabel('Energy sources')
plt.ylabel('Energy consumption (MWh)')
plt.show()
```

## Pie Chart showing the share of each Renewable Energy source in the Total Energy Consumption

```python
# Pie chart to show energy consumption distribution
plt.pie(energy_values, labels=energy_sources, autopct='%1.1f%%', colors=['gold', 'lightblue', 'lightgreen', 'coral'])

# Add title
plt.title('Energy Consumption Share by Source')
plt.show()
```

## Scatter Plot for Analyzing the Relationship Between Energy Consumption and Carbon Emissions

```python
# Sample data for energy consumption and carbon emissions
carbon_emissions = [200, 500, 450, 300] # in Kg CO2

# Create a scatter plot
plt.scatter(energy_values, carbon_emissions, color='red')

# Add labels and title
plt.title('Energy Consumption vs Carbon Emissions')
plt.xlabel('Energy Consumption (MWh)')
plt.ylabel('Carbon Emissions (Kg CO2)')
plt.show()
```

## Customizing a Bar Chart

```python
# Create a customized bar chart
plt.bar(energy_sources, energy_values, color='orange', edgecolor='black', linewidth=2)

# Add labels and title
plt.title('Customized Energy Consumption by Sources', fontsize=14, fontweight='bold')
plt.xlabel('Energy Sources', fontsize=12)
plt.ylabel('Energy Consumption (MWh)', fontsize=12)

# Adding grid and customizing style
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend(['Energy Consumption'], loc='upper right')
plt.show()
```

## Pair Plot

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# Creating a DataFrame for pair plot
data = {
    "Solar": [1200, 1500, 1300],
    "Wind": [3400, 3600, 3200],
    "Hydropower": [2900, 3100, 2800],
    "Biomass": [2500, 2700, 2400]
}
df = pd.DataFrame(data)

# Create a pair plot
sns.pairplot(df)
plt.show()
```

## Heatmap Showing the Correlation Between Different types

```python
# Create a heatmap to show the correlation matrix
correlation_matrix = df.corr()

# plot heatmap
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)

# add title
plt.title('Correlation Between Energy Sources')
plt.show()
```

For Heat map:
- -0.7 to -1: Strong negative correlation
- -0.5 to -0.7: Moderate negative correlation
- -0.3 to -0.5: Weak negative correlation
- 0: No correlation
- 0.3 to 0.5: Weak positive correlation
- 0.5 to 0.7: Moderate positive correlation
- 0.7 to 1: Strong positive correlation

## Box Plot Comparing the Distribution of Energy By Source

```python
# Create a box plot
sns.boxplot(data=df)

# Add title
plt.title('Distributuion of Energy Consumption by Source')
plt.show()
```

## Violin Plot Comparing Energy Consumption Distributions

```python
# Create a violin plot
sns.violinplot(data=df)

# Add title
plt.title('Violin Plot of Energy Consumption Distributions')
plt.show()
```

## Regression Plot

```python
# Sample data for energy consumption and carbon emissions
energy_values = [100, 200, 300, 400, 500]
carbon_emissions = [10, 20, 30, 40, 50]

# Create a DataFrame with energy consumption and carbon emission
df_reg = pd.DataFrame({'Energy Consumption': energy_values, 'Carbon Emissions': carbon_emissions})

# Create a regeression plot
sns.regplot(x='Energy Consumption', y='Carbon Emissions', data=df_reg)

# Add title
plt.title('Energy Consumption ys Carbon Emissions')
plt.xlabel('Energy Consumption')
plt.ylabel('Carbon Emissions')
plt.show()
```

## Facet Grid

```python
# Define the months
months = ['January', 'February', 'March', 'April', 'May', 'June']

# Creating a dataset for different regions and energy consumption over months
df_facet = pd.DataFrame({
    'Month': months * 3,
    'Energy Consumption': [1200, 1300, 1100, 1500, 1400, 1600] * 3,
    'Region': ['North', 'South', 'East'] * 6
})

# Create a FacetGrid plot
g = sns.FacetGrid(df_facet, col='Region', hue='Region', height=4, aspect=1)
g.map(sns.lineplot, 'Month', 'Energy Consumption')

# Add title
plt.subplots_adjust(top=0.85)
g.fig.suptitle('Energy Consumption Over Time by Region')
plt.show()
```
